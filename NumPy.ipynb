```python
import numpy as np

data = np.genfromtxt('Advertising.csv',delimiter = ',', skip_header = 1)

print(data.shape)
print(data[:5])
```

```python
import pandas as pd
df = pd.read_csv('Advertising.csv')
df
```

```python
# Find the mean, median, and standard deviation of the Sales column.
sales = data[:,4]
mean = np.mean(sales)
print(mean)

median = np.median(sales)
print(median)

sd = np.std(sales)
print(sd)
```

```python
# Which advertising channel (TV, Radio, Newspaper) has the highest average spending?
tv = data[:,1]
tv_mean = np.mean(tv)
radio = data[:,2]
radio_mean = np.mean(radio)
newspaper = data[:,3]
newspaper_mean = np.nanmean(newspaper)

if tv_mean > radio_mean and tv_mean > newspaper_mean:
  print("tv has higest average spending",tv_mean)
elif radio_mean > tv_mean and radio_mean > newspaper_mean:
  print("radio has highest average spending",radio_mean)
else:
  print("Newspaper has highest average spending",newspaper_mean)
```

```python
# Count how many campaigns have Sales greater than 20.
sales_count = data[:,4]
mask = sales_count > 20
print(data[mask])
print(np.sum(mask))
```

```python
# Extract all rows where:

# TV > 200

# Sales > 15

# Return the number of such campaigns.

tv = data[:,1]
sales = data[:,4]

mask1 = tv > 200
mask2 = sales > 15
ans = mask1 & mask2
print(data[ans])
print(np.sum(ans))
# print(np.array([ans]))
```

```python
sales = data[:,4]
tv = data[:,1]
radio = data[:,2]
newspaper = data[:,3]
effeciency = sales / (tv + radio + newspaper)
idx = np.argsort(effeciency)
top5_idx = idx[-5:]
print(data[top5_idx])
```

```python
# Calculate the correlation coefficient between:

# TV & Sales

# Radio & Sales

# Newspaper & Sales
tv = data[:,1]
sales = data[:,4]
radio = data[:,2]
newspaper = data[:,3]
mask = ~np.isnan(tv) & ~np.isnan(sales) & ~np.isnan(radio) & ~np.isnan(newspaper)
tv_clean = tv[mask]
sales_clean = sales[mask]
radio_clean = radio[mask]
newspaper_clean = newspaper[mask]
corr_tv_sales = np.corrcoef(tv_clean,sales_clean)[0,1]

corr_radio_sales = np.corrcoef(radio_clean,sales_clean)[0,1]

corr_newspaper_sales = np.corrcoef(newspaper_clean, sales_clean)[0,1]

corr = np.array([corr_tv_sales,corr_radio_sales,corr_newspaper_sales])

channels = np.array(["TV","Radio","Newspaper"])

idx = np.argmax(corr)

print(channels[idx] , "has the highest correlation with sales: ", corr[idx])
```

```python
# Normalization challenge
# Normalize TV, Radio, and Newspaper columns using min-max normalization:
# (x - min) / (max - min)
# Return a new NumPy array with normalized values.

tv = data[:,1]
radio = data[:,2]
newspaper = data[:,3]

tv_min = np.nanmin(tv)
tv_max = np.nanmax(tv)
radio_min = np.nanmin(radio)
radio_max = np.nanmax(radio)
newspaper_min = np.nanmin(newspaper)
newspaper_max = np.nanmax(newspaper)

tv_norm = (tv - tv_min)/(tv_max - tv_min)
radio_norm = (radio - radio_min)/(radio_max - radio_min)
newspaper_norm = (newspaper - newspaper_min)/(newspaper_max - newspaper_min)

norm_val = np.column_stack([tv_norm,radio_norm,newspaper_norm])
norm_val
```

```python
# Find campaigns where Sales is more than 2 standard deviations above the mean.
# Return:
# Number of such campaigns
# Their Sales values

sales = data[:,4]
sales_std = np.nanstd(sales)

mask =  sales > np.nanmean(sales) + 2 * sales_std
outliers = sales[mask]

print(outliers)
print(np.sum(mask))
```

```python
# Find campaigns where:

# TV spending is above average

# BUT Sales are below average

tv = data[:,1]
sales = data[:,4]

tv_mean = np.nanmean(tv)
sales_mean = np.nanmean(sales)

mask = (tv > tv_mean) & (sales < sales_mean)

print(np.sum(mask))
```

```python
# Compare:
# 1️⃣ Average Sales when both TV and Radio spending are high
# 2️⃣ Overall average Sales

tv = data[:,1]
radio = data[:,2]
tv_avg = np.nanmean(tv)
radio_avg = np.nanmean(radio)
sales = data[:,4]
avg_sales = (tv > tv_avg) & (radio > radio_avg)
print(sales[avg_sales])
np.nanmean(sales[avg_sales])
```

```python
# Q1️⃣ Batting overview
# Compute:
# Mean, median, std of Runs
# Mean Strike Rate

runs = data[:,5]
strike_rate = data[:,9]
mean = np.nanmean(runs)
median = np.nanmedian(runs)
std = np.nanstd(runs)
mean_SR = np.nanmean(strike_rate)
print(mean,median,std,mean_SR)
```

```python
# How many players have Runs greater than the average Runs?

runs = data[:,5]
avg_runs = np.nanmean(runs)
players = runs > avg_runs
print(np.sum(players))
```

```python
# Count how many players satisfy both conditions:
# Innings ≥ 20
# AND
# Average ≥ 30

Innings = data[:,3]
average = data[:,8]

players = (Innings >= 20) & (average >=30)
print(np.sum(players))
```

```python
# Which players score runs efficiently WITHOUT relying on boundaries?

runs = data[:,5]
balls_faced = data[:,6]
strike_rate = data[:,9]

runs_per_ball = runs / balls_faced

mask = (runs_per_ball >= np.median(runs_per_ball)) & (strike_rate >= np.median(strike_rate))

print(np.sum(mask))
```

```python
runs = data[:,5]
strike_rate = data[:,9]

players = (runs > np.median(runs)) & (strike_rate > np.median(strike_rate))

print(np.sum(players))
print(runs[players])
print(strike_rate[players])
```

```python
# Detect players where:
# Runs > mean(Runs) + 2 × std(Runs)

mask = runs > np.nanmean(runs) + 2 * np.nanstd(runs)
print(np.sum(mask))
print(runs[mask])
print(strike_rate[mask])
```

```python
# Find the top 5 players with highest efficiency using np.argsort.
idx = np.argsort(runs_per_ball)
top5_idx = idx[-5:][::-1]
print("runs per ball: ",runs_per_ball[top5_idx])
print("Runs: ",runs[top5_idx])
print("strike rate: ", strike_rate[top5_idx])
```

```python
# Which players faced more balls than average but still scored fewer runs than average?

balls_faced = data[:,6]
runs = data[:,5]

mask = (balls_faced > np.nanmean(balls_faced)) & (runs < np.nanmean(runs))
print(np.sum(mask))
print("runs: ", runs[mask])
print("Strike rate: ", strike_rate[mask])
```

```python
# Min–max normalize:
# Runs
# Balls
# Strike Rate

runs = data[:,5]
balls_faced = data[:,6]
strike_rate = data[:,9]

norm_runs = (runs - np.nanmin(runs)) / (np.nanmax(runs) - np.nanmin(runs))
norm_balls_faced = (balls_faced - np.nanmin(balls_faced)) / (np.nanmax(balls_faced) - np.nanmin(balls_faced))
norm_strike_rate = (strike_rate - np.nanmin(strike_rate)) / (np.nanmax(strike_rate) - np.nanmin(strike_rate))

total_norm = np.column_stack((norm_runs,norm_balls_faced,norm_strike_rate))
print(total_norm)
total_norm.shape
```

```python
# Compare average Runs for players where:
# Strike Rate > median(Strike Rate)
# AND
# Runs > median(Runs)

strike_rate = data[:,9]
runs = data[:,5]

mask = (strike_rate > np.nanmedian(strike_rate)) & (runs > np.nanmedian(runs))

filtered_avg_runs = np.nanmean(runs[mask])
overall_avg_runs = np.nanmean(runs)



print(filtered_avg_runs)
print(overall_avg_runs)
```
